# Correlation 
**Definition: Correlation measures the strength and direction of a relationship between two variables.**

- +1: Strong positive correlation
- 0: No correlation
- â€“1: Strong negative correlation

## Types of Correlation in Python with Pandas

- df.corr(method='pearson')   # default
- df.corr(method='spearman')  # good for ranked/nonlinear data
- df.corr(method='kendall')   # more robust for small datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("penguins") 
df.head()
```

```python
corr_matrix = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot = True  ,cmap = 'coolwarm' , fmt = '.2f')
plt.title("Correlation Matrix of Penguins Dataset")
plt.show()
```
