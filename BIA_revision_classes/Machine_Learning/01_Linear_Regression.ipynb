- Linear Regression is a supervised machine learning algorithm used to predict a numeric value (output) based on one or more input variables (features), by finding the best-fitting straight line through the data.
  
- Formula :  Y = mX + b

| Symbol | Meaning                            |
| ------ | ---------------------------------- |
| Y      | Predicted value (target)           |
| X      | Input value (feature)              |
| m      | Slope of the line                  |
| b      | Intercept (where line hits Y-axis) |


| Type           | Description                                                       |
| -------------- | ----------------------------------------------------------------- |
| **Simple**     | One feature (X) and one target (Y)                                |
| **Multiple**   | Multiple features (X‚ÇÅ, X‚ÇÇ, X‚ÇÉ...) and one target (Y)              |
| **Polynomial** | Not a straight line but curved ‚Äî still based on linear principles |

# <a id='toc2_'></a>[Step-01 Load Libraries](#toc0_)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

# <a id='toc3_'></a>[Step-02 Data Preparation](#toc0_)

```python
np.random.seed(0)
hours_studied = np.random.uniform(0,10,100)
exam_scores = 50 + 5*hours_studied + np.random.normal(0,2,100)
```

```python
print(f"Hours_studies: {len(hours_studied)}")
print(f"exam_scores: {len(exam_scores)}")
```

```python
df = pd.DataFrame({"Hours_studied":hours_studied , "Exam_scores":exam_scores})
df.head(20)
```

# <a id='toc4_'></a>[Step-03 Splitting the data](#toc0_)

```python
df.columns
```

```python
X = df[["Hours_studied"]]
y = df["Exam_scores"]

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2, random_state=32) # shuffle the data

print(len(X_train), len(X_test) , len(y_train), len(y_test))
```

```python
X_train
```

# <a id='toc5_'></a>[Step-04 Model Selection](#toc0_)

```python
model_lr = LinearRegression()
model_lr.fit(X_train , y_train)
```

# <a id='toc6_'></a>[Step-05 Model Prediction](#toc0_)


```python
y_pred = model_lr.predict(X_test)
```

# <a id='toc7_'></a>[Step-06 Model Evaluation](#toc0_)

```python
mse = mean_squared_error(y_test , y_pred) # difference between Actual and Predicted values
r2 = r2_score(y_test,y_pred)

print(f"mse: {mse:.2f}")
print(f"R_squared: {r2:.2f}")
```

# <a id='toc8_'></a>[Step-07 Create Dataframe](#toc0_)

```python
dataframe_lr = pd.DataFrame({"Actual_scores":y_test , "Predicted_scores":y_pred})
dataframe_lr.head()
```

```python
X_test['y_test'] = y_test
X_test['y_pred'] = y_pred
display(X_test)
```

# <a id='toc9_'></a>[Step-09 Data Visualization and Interpretation](#toc0_)

```python
import matplotlib.pyplot as plt

# Scatter plot of actual test data points
plt.scatter(X_test, y_test, color='blue', label='Actual Data')
# Plotting the regression line
plt.plot(X_test, y_pred, color='red', label='Regression Line', linewidth=2)
# Label the x-axis
plt.xlabel('Hours Studied')
# Label the y-axis
plt.ylabel('Exam Scores')
# Add a legend to the plot
plt.legend()
# Set the title of the plot
plt.title('Simple Linear Regression: Hours Studied vs. Exam Scores')
# Display the plot
plt.show()
```

```python
# Extract the slope (coefficient) and intercept from the trained model
slope = model_lr.coef_[0]
intercept = model_lr.intercept_

# Print the slope and intercept with formatting
print(f'Slope (Coefficient): {slope:.2f}')
print(f'Intercept: {intercept:.2f}')
```
